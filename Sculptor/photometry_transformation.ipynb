```python
from sympy import *
init_printing()
```

```python
bands = ["g","r","i","z","y"]
g_sdss,r_sdss,i_sdss,z_sdss,y_sdss = symbols(" ".join([band+"_sdss" for band in bands]))
g_des,r_des,i_des,z_des,y_des = symbols(" ".join([band+"_des" for band in bands]))
```

```python
des_to_sdss = solve([
    Eq(g_des , g_sdss - 0.104*(g_sdss-r_sdss) + 0.01),
    Eq(r_des , r_sdss - 0.102*(g_sdss-r_sdss) + 0.02),
    Eq(i_des , i_sdss - 0.256*(i_sdss-z_sdss) + 0.02),
    Eq(z_des , z_sdss - 0.086*(i_sdss-z_sdss) + 0.01),
    Eq(y_des , y_sdss + 0.238*(z_sdss-y_sdss) + 0.483)
],[g_sdss,r_sdss,i_sdss,z_sdss,y_sdss])
des_to_sdss
```

```python
for band in g_sdss,r_sdss,i_sdss,z_sdss,y_sdss:
    print('ret["{}"]'.format(band),"=",str(N(des_to_sdss[band],4)))
```

```python
# for example,
example = {g_des:20,i_des:22,r_des:10}
for var,eq in des_to_sdss.items():
    display({var:eq.subs(example)})
```
