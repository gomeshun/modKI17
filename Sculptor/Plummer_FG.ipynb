```python
import matplotlib.pyplot as plt
from sympy import *
init_printing()
```

```python
R,re = symbols("R r_e",positive=True)
```

```python
plummer = (1/pi/re**2)*(1+(R/re)**2)**(-2)
plummer,plummer.subs(R,re)
```

```python
odds = 40
flat = 1/odds*plummer.subs(R,re)
RoI_R = 3750.0
```

```python
f1 = (flat*(2*pi*R)).subs({re:226})
f2 = (plummer*(2*pi*R)).subs({re:226})
norm = simplify(integrate(f1,R) + integrate(f2,R))
norm = N(norm.subs(R,RoI_R) - norm.subs(R,0))
norm
```

```python
pf = plot(f1/norm,f2/norm,(f1+f2)/norm,(R,0,RoI_R),show=False)
pf[0].line_color = "C1"
pf[1].line_color = "C0"
pf[2].line_color = "black"
pf.show()
#pf2 = plot(f2/norm,(R,0,RoI_R),line_color="C0")
#pf3 = plot((f1+f2)/norm,(R,0,RoI_R),line_color="black")
#pf2.append(pf1[0])
#pf2.append(pf3[0])
#pf2.show()
```
