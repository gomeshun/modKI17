```python
import pandas as pd
```

```python
dsph = pd.read_csv("dsphdata.csv",index_col=0)
dsph
```

```python
dsph_new = pd.DataFrame()
dsph_new["name"] = dsph.index

# hms
rah,ram,ras = [],[],[] 
for s in dsph["RA(J2000)[hh:mm:ss]"]:
    tmp = s.split(":")
    rah.append(tmp[0])
    ram.append(tmp[1])
    ras.append(tmp[2])
dsph_new["RAd"],dsph_new["RAm"],dsph_new["RAs"] = rah,ram,ras
    
# dms
ded,dem,des = [],[],[] 
for s in dsph["DEC(J2000)[dd:mm:ss]"]:
    tmp = s.split(":")
    ded.append(tmp[0].strip("$"))
    dem.append(tmp[1])
    des.append(tmp[2])
dsph_new["DEd"],dsph_new["DEm"],dsph_new["DEs"] = ded,dem,des

dsph_new.DEd
```
