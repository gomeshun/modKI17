```python
import pandas as pd
import Calc_Jfactor_w_likeli as calcJ
import numpy as np
```

```python
fname = "MCMC_spectroscopy_sig_refR-re-w-center_2gaussian_plummer_Rsig-3000pc_emcee_v002.csv"
out_fname = "draco_mcmc.df"
```

```python
# load data
df = pd.read_csv(fname,index_col=0)
```

```python
display(df.columns)
rhos_Msunpc,rs_pc,a,b,g,dist = df.log10_rhos_Msunpc3, df.log10_rs_pc, df.a, df.b, df.g, df.dist
rhos_Msunpc,rs_pc = np.power(10,rhos_Msunpc), np.power(10,rs_pc)
```

```python
J = calcJ.Jfactor_v02(rhos_Msunpc,rs_pc,a,b,g,dist,Rtrunc_pc=2000)
```

```python
df["log10_J_GeV2cm5"] = np.log10(J)
```

```python
comment = (
"""# MCMC dataset generated by emcee.
# ---------------- MCMC info ---------------- #
# Generated Date: 2018/12/01, in Taiwan
# 
# 
# dSph: Draco 
#     profile: Plummer+NFW
# FG: 
#     profile: 2 gaussian distribution
#     
# Signal Resion: R < 3000 pc
# Control Resion: R > 3000 pc
# 
""")
file = open(out_fname,mode="w")
file.write(comment)
file.close()
df.to_csv(out_fname,header=comment,mode="a")
```
