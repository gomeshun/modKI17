```python
besfilename = "Besancon/1538022407.981358.resu"
```

```python
f = open(besfilename)
ftxt = f.readlines()
fstart = 95
fend = -6
data = ftxt[fstart:fend]
f.close()

f = open(besfilename+".csv",mode="w")
[ f.write(s) for s in data ]
f.close()

f = open(besfilename+".head.csv",mode="w")
[ f.write(s) for s in data[0:1] ]
f.close()

f = open(besfilename+".csv")
print(f.read())
f.close()
```

```python
bescsvname = besfilename + ".csv"
besindexcsvname = besfilename + ".head.csv"
npindex = np.loadtxt(besindexcsvname,dtype=str)
npdata = np.loadtxt(bescsvname,skiprows=1)
df = pd.DataFrame(npdata)
df.columns = npindex
display(df)
df.to_csv("Besancon.csv")
```

```python
df.plot.scatter("V-I","V",s=1)
plt.ylim(20,15)
plt.xlim(0,2)
```
