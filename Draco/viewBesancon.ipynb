```python
besfilename = "Besancon/draco_bes.resu.csv"
```

```python
df = pd.read_csv(besfilename,delim_whitespace=True)
df
```

```python
display(df.columns)
df.plot.scatter("i","Vr",s=1,figsize=(8,6))
df.plot.scatter("g-i","i",s=1,figsize=(8,6))
df.plot.scatter("g-i","Vr",s=1,figsize=(8,6))
#plt.ylim(20,15)
#plt.xlim(0,2)
```

```python
df["g-i"] = df["g-r"] + df["r-i"]
```

```python
df.plot.scatter("g-i","i",s=1)
plt.ylim([22,14])
plt.xlim([-0.5,2.5])
```

```python
from cmd_utilities import inpoly
cut = np.loadtxt("Draco_CMDcut.csv",delimiter=",")
cut
```

```python
df_filtered = df[inpoly(x=cut[:,0],y=cut[:,1],sx=df["g-i"].values,sy=df["i"].values)]
```

```python
df_filtered.plot.scatter("g-i","i",s=1)
plt.ylim([22,14])
plt.xlim([-0.5,2.5])
len(df_filtered)
```

```python
bins = np.linspace(-500,200,64)
df_filtered.Vr.hist(alpha=0.3,bins=bins,label="tot")
df_filtered[df_filtered["[Fe/H]"]>-0.5].Vr.hist(alpha=0.5,bins=bins,figsize=(8,6),label="thin: -0.5<[Fe/H]")
df_filtered[(df_filtered["[Fe/H]"]<=-0.5) & (df_filtered["[Fe/H]"]>-1.2)].Vr.hist(alpha=0.5,bins=bins,label="thick:-1.2<[Fe/H]<0.5")
df_filtered[(df_filtered["[Fe/H]"]<=-1.2)].Vr.hist(alpha=0.5,bins=bins,label="halo:[Fe/H]<-1.2")
df_filtered[df_filtered["[Fe/H]"]<=-2.2].Vr.hist(alpha=0.5,bins=bins,figsize=(8,6))
#plt.yscale("log")
plt.title("Besancon model (after color-magnitude cut)")
plt.xlabel(r"$v_{los}$[km/s]")
plt.legend()
plt.text(110,45,s=r"$N_\mathrm{}={}$".format("{plot}",len(df_filtered)))
```

```python
bins = np.linspace(-500,200,64)
df.Vr.hist(alpha=0.3,bins=bins,label="tot")
df[df["[Fe/H]"]>-0.5].Vr.hist(alpha=0.5,bins=bins,figsize=(8,6),label="thin: -0.5<[Fe/H]")
df[(df["[Fe/H]"]<=-0.5) & (df["[Fe/H]"]>-1.2)].Vr.hist(alpha=0.5,bins=bins,label="thick:-1.2<[Fe/H]<0.5")
df[(df["[Fe/H]"]<=-1.2)].Vr.hist(alpha=0.5,bins=bins,label="halo:[Fe/H]<-1.2")
#df_filtered[df_filtered["[Fe/H]"]<=-2.2].Vr.hist(alpha=0.5,bins=bins,figsize=(8,6))
#plt.yscale("log")
plt.title("Besancon model (before color-magnitude cut)")
plt.xlabel(r"$v_{los}$[km/s]")
plt.legend()
plt.text(110,1100,s=r"$N_\mathrm{}={}$".format("{plot}",len(df)))
```
